```python
import numpy as np
import cv2 as cv
import matplotlib.pyplot as plt
```

```python
img_path = './tamagoyaki/tamagoyaki_0000.jpeg'
img = cv.imread(img_path)
```

```python
img.shape
```

```python
plt.imshow(img)
```

```python
plt.imshow(cv.cvtColor(img, cv.COLOR_BGR2RGB))
```

```python
fix_img = cv.resize(img, (64, 64))
```

```python
plt.imshow(cv.cvtColor(fix_img, cv.COLOR_BGR2RGB))
```

```python
cv.imwrite('./fix_tamagoyaki/fix_tamagoyaki.jpg', fix_img)
```
